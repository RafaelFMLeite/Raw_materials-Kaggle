# Importando bibliotecas

```python
import pandas as pd
import numpy as np
```

##Leitura de arquivo

```python
df = pd.read_csv("agricultural_raw_material.csv")
```

## Informações do dataframe

```python
display(df)
print(df.info())
```

```python
df.isnull().sum()
```

## Tratamento de dados

```python
df = df.replace('%', '', regex=True)
df = df.replace(',','', regex=True)
df = df.replace('-', '', regex=True)
#NaN Not a number, lembrando que NaN é diferente de null
df = df.replace('', np.nan)
#Nesta parte do código, eu estou apagando todos os valores NaN encontrados no dataframe
df = df.dropna()
df.isnull().sum()
#Check dos dados nulos
```

```python
#Convertendo todas as colunas  para float
lst = ['Coarse wool Price', 'Coarse wool price % Change', 'Copra Price', 'Copra price % Change', 'Cotton Price', 'Cotton price % Change', 'Fine wool Price',
       'Fine wool price % Change', 'Hard log Price', 'Hard log price % Change', 'Hard sawnwood Price',
       'Hard sawnwood price % Change','Hide Price', 'Hide price % change', 'Plywood Price', 'Plywood price % Change', 'Rubber Price', 'Rubber price % Change',
       'Softlog Price', 'Softlog price % Change', 'Soft sawnwood Price', 'Soft sawnwood price % Change', 'Wood pulp Price', 'Wood pulp price % Change']

df[lst] = df[lst].astype(float)

#Visualização das colunas tratadas
df.dtypes
```

```python
#Veja como os dados estão mais simples de serem visualizados agora
df.head()
```

```python
#Agora o tratamento de dados irá ser para a data, onde mais tarde poderemos visualizar o data series
df.Month = pd.to_datetime(df.Month, format='%b%y', yearfirst =False)

#colocando Month como o indice da tabela
df = df.set_index('Month')
df.head()
```

## Analise exploratória usando matplotlib e seaborn

```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
sns.set_style('whitegrid')
```

Mapa de calor da correlação dos preços

```python
raw_data = ['Copra Price','Coarse wool Price', 'Cotton Price', 'Fine wool Price', 'Hard log Price', 'Hard sawnwood Price', 'Plywood Price',
            'Rubber Price', 'Softlog Price', 'Soft sawnwood Price', 'Wood pulp Price']
#correlação entre os preços dos produtos de materia prima
correlation = df[raw_data].corr()
mask = np.triu(np.ones_like(correlation))
plt.figure(figsize=(12, 9))
sns.heatmap(correlation, mask=mask ,annot=True, cmap='coolwarm')
plt.show()
```



*   Pode-se analisar com este insight que a correlação entre cada materia prima, sendo maior o valor, maior a correlação entre elas, mas não necessariamente

Mapa de calor da correlação do percentual de variação de preços de materias primas

```python
change_list = ['Copra price % Change','Softlog price % Change','Rubber price % Change','Cotton price % Change','Coarse wool price % Change','Fine wool price % Change','Hard log price % Change','Hard sawnwood price % Change','Hide price % change','Plywood price % Change','Soft sawnwood price % Change','Wood pulp price % Change']
correlation_porcentage_price = df[change_list].corr()
mask = np.triu(np.ones_like(correlation_porcentage_price))
plt.figure(figsize=(12, 9))
sns.heatmap(correlation_porcentage_price, mask=mask, annot=True,cmap='Blues')
```

**Variação percentual das matérias ao longo dos anos**

Variação percentual das matérias

```python
import random as random
change_list = ['Copra price % Change','Softlog price % Change','Rubber price % Change','Cotton price % Change','Coarse wool price % Change','Fine wool price % Change','Hard log price % Change','Hard sawnwood price % Change','Hide price % change','Plywood price % Change','Soft sawnwood price % Change','Wood pulp price % Change']

plt.figure(figsize=(15,12))

for i in range(len(change_list)):
  r = random.random()
  g = random.random()
  b = random.random()
  color = (r,g,b)
  plt.subplot(4,3,i+1)
  plt.subplots_adjust(hspace=1, wspace=0.3)
  plt.plot(df[change_list[i]],c=color)
  plt.ylabel('Variação %')
  plt.xticks(rotation=90)
  plt.legend(change_list[i:],loc='upper center', bbox_to_anchor=(0.5,1.4),fontsize=10)
```



*   Menor variação percentual compensada, 20
*   Maior variação percentual em madeira serrada, 60
